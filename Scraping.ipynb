```python
import requests
from bs4 import BeautifulSoup
```

```python
url='https://www.carrefour.es/neumaticos/cat3113400/c'
```

```python
ruedas_carrefour=requests.get(url)
print(ruedas_carrefour.text)
```

```python
soup = BeautifulSoup(ruedas_carrefour.content, 'html.parser')
soup.title.text
```

```python
ofertas=soup.find_all(class_='product-card')
```

```python
lista_ofertas = []
for oferta in ofertas:
    rueda=oferta.find('h2').text.strip()
    precio_corriente=oferta.find('span').text.strip() 
    lista_ofertas.append([rueda,precio_corriente])
lista_ofertas
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(lista_ofertas,columns=['rueda','precio_corriente'])
df
```

```python
url='http://emojitracker.com'
```

```python
emoticones=requests.get(url)
```

```python
soup = BeautifulSoup(emoticones.content, 'html.parser')
soup.title.text
```

```python
emoticones_lista=soup.ul.il.string
emoticones_lista
```

```python
lista_emoji=[]
for emoji in emoticones_lista:
    titulo =emoji.li.get('data-title')
    print(titulo)
    lista_emoji.append(titulo)
    
lista_emoji
    
    
```
