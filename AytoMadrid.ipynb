```python
import requests
```

```python
url ='https://datos.madrid.es/egob/catalogo/206974-0-agenda-eventos-culturales-100.json'
```

```python
response = requests.get(url)
response.status_code
```

```python
response.json()
```

```python
#visulaizar el primer item
response.json()['@graph'][0]
```

```python
#Calcular numero de actividades
n_titulos=len(response.json()['@graph'])
f'El número de eventos culturales en los próximos 100 días es: {n_titulos}'
```

```python
for i in range(n_titulos):
    titulos=response.json()['@graph'][i]['title']
    print(titulos)
```

```python
#Donde podemos asistir a ver la pelicula "Bienvenido Mr. Marshall" y que día y hora

for i in range(n_titulos):
    titulo=response.json()['@graph'][i]['title']
    if titulo =='Bienvenido Mr. Marshall':
        print(f'El item de Bienvenido Mr. Marshall es: {i}')
   
```

```python
item=63
localizacion = response.json()['@graph'][item]['address']['area']['street-address']
hora= response.json()['@graph'][item]['time']
fecha=response.json()['@graph'][item]['dtstart'].split(' ')[0].split('-')
f = "/".join(reversed(fecha))

f'"Bienvenido Mr. Marshal" sera proyectada el {f} a las {hora} en {localizacion}'
```
